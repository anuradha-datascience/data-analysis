#### Purpose: Analyze how IMDb scores vary across different film genres.

```python
# Determine the top 10 film genres by frequency
top_10_genres = df['Genre'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 genres
df_top_10_genres = df[df['Genre'].isin(top_10_genres)]

# Calculate the mean IMDb score for each genre
genre_means = df_top_10_genres.groupby('Genre')['IMDB Score'].mean().reset_index()

# Sort the genres by mean IMDb score in descending order
genre_means = genre_means.sort_values(by='IMDB Score', ascending=False)


# Create a bar plot of mean IMDb scores for the top 10 film genres

sns.barplot(data=genre_means, x='Genre', y='IMDB Score', color='red')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Set plot title and labels
plt.title('Mean IMDb Scores by Top 10 Film Genres')
plt.xlabel('Film Genre')
plt.ylabel('Mean IMDb Score')

# Show the plot
plt.show()

```

#### Purpose: Understand the evolution of genre distribution over the years.

The resulting heatmap visually represents how the distribution of film genres has evolved over the years. Each cell in the heatmap corresponds to a specific year and genre, and the color intensity indicates the frequency of that genre's appearance in films released in that year. 

```python
# Genre Distribution Over Time
# Determine the top 10 film genres by frequency
top_10_genres = df['Genre'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 genres
df_top_10_genres = df[df['Genre'].isin(top_10_genres)]

# Create a pivot table for the top 10 genres and their distribution over time
genre_pivot = df_top_10_genres.pivot_table(index='Premiere_Year', columns='Genre', aggfunc='size', fill_value=0)

# Create a heatmap
# plt.figure(figsize=(12, 6))
#sns.heatmap(genre_pivot, cmap='coolwarm')
sns.heatmap(genre_pivot, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5)

plt.title('Top 10 Genre Distribution Over Time')
plt.xlabel('Genre')
plt.ylabel('Year of Premiere')
plt.show()
```

```python


# Determine the top 10 film genres by frequency
top_10_genres = df['Genre'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 genres
df_top_10_genres = df[df['Genre'].isin(top_10_genres)]

# Group the data by year and genre and count the number of films in each group
genre_counts = df_top_10_genres.groupby(['Premiere_Year', 'Genre']).size().reset_index(name='Film_Count')

# Create a line chart to visualize the distribution over time
# plt.figure(figsize=(12, 6))

    
for genre in top_10_genres:
    genre_data = genre_counts[genre_counts['Genre'] == genre]
    plt.plot(genre_data['Premiere_Year'], genre_data['Film_Count'], label=None)  # No label here
    last_data_point = genre_data.iloc[-1]  # Get the last data point for labeling
    va = 'bottom' if genre_data['Film_Count'].iloc[-1] < 100 else 'center'
    plt.text(last_data_point['Premiere_Year'], last_data_point['Film_Count'], genre, ha='left', va=va)

# Set plot title and labels
plt.title('Top 10 Genre Distribution Over Time')
plt.xlabel('Year of Premiere')
plt.ylabel('Number of Films')

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Language Distribution Over Time
language_pivot = df.pivot_table(index='Premiere_Year', columns='Language', aggfunc='size', fill_value=0)
# plt.figure(figsize=(12, 6))
sns.heatmap(language_pivot, cmap='coolwarm')
plt.title('Language Distribution Over Time')
plt.xlabel('Language')
plt.ylabel('Year of Premiere')
plt.show()
```

```python

# plt.figure(figsize=(12, 6))

month_counts = df.groupby('Premiere_Month').size().reset_index(name='Count')
month_counts


# Find the month(s) with the highest count
highest_months = month_counts[month_counts['Count'] == month_counts['Count'].max()]['Premiere_Month'].tolist()

# Create a custom color palette
colors = ['red' if month in highest_months else 'gray' for month in month_counts['Premiere_Month']]

# Create the countplot with the custom colors
plt.figure(figsize=(12, 6))
sns.countplot(x='Premiere_Month', data=df, palette=colors)
plt.title('Release Month Analysis')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
